```python
import os
import pandas as pd
from ebmdatalab import bq
```

```python
sql = f"""
WITH
  list AS (
  FROM
    hscic.practice_statistics 
  INNER JOIN
    ebmdatalab.hscic.practices prac
  ON
    practice = prac.code
    AND setting = 4
    AND
    month >= '2020-01-01')
SELECT
    month,
    total_list_size, 
    pct_id,
    practice, 
    name, 
    ccg_id AS sicbl, 
    pcn_id, 
    status_code
FROM
   list
"""

# Define the CSV path for caching the results.
csv_path = os.path.join('..', 'data', 'practice_info.csv.gz')

# Use the cached_read function from the bq library to run the query.
vtm_matched = bq.cached_read(sql, csv_path=csv_path)
```
